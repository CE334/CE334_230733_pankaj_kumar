## **13. Derivative Function**

```python
#chat gpt help had been taken to construct the function and compare the analytical values with numerical values
import math  #standered math library had been imported to implement mathematical function and operations

def derivative(f, a, method, h=0.01):
    """
    Compute the numerical derivative of function f at point a.

    Parameters:
    f      : function handle (e.g., math.exp)
    a      : point at which to compute the derivative
    method : 'forward', 'backward', or 'central' (default: 'central')
    h      : step size (default: 0.01)
    """
    if method == 'forward':
        return (f(a + h) - f(a)) / h
    elif method == 'backward':
        return (f(a) - f(a - h)) / h
    elif method == 'central':
        return (f(a + h) - f(a - h)) / (2 * h)
    else:
        raise ValueError("Method must be 'forward', 'backward', or 'central'")

# --- Testing the Function ---

x_val = 0.5

# Test 1: f(x) = e^x
# Analytical derivative of e^x is e^x. At x=0.5, exact value is e^0.5
exact_exp = math.exp(x_val)

print(f"--- Derivative of e^x at x = {x_val} ---") # here python f string is used to simplyfy the print operation
print(f"Exact Value: {exact_exp:.6f}")  # .6f will print 6 integer after decimal on output screen
print(f"Forward:     {derivative(math.exp, x_val, 'forward', h=0.01):.6f}") #------|
print(f"Backward:    {derivative(math.exp, x_val, 'backward', h=0.01):.6f}") #     |--numerical derivatives
print(f"Central:     {derivative(math.exp, x_val, 'central', h=0.01):.6f}") #------|

print("\n")

# Test 2: f(x) = cos(x)
# Analytical derivative of cos(x) is -sin(x). At x=0.5, exact value is -sin(0.5)
exact_cos = -math.sin(x_val)

print(f"--- Derivative of cos(x) at x = {x_val} ---")
print(f"Exact Value: {exact_cos:.6f}")
print(f"Forward:     {derivative(math.cos, x_val, 'forward', h=0.01):.6f}")
print(f"Backward:    {derivative(math.cos, x_val, 'backward', h=0.01):.6f}")
print(f"Central:     {derivative(math.cos, x_val, 'central', h=0.01):.6f}")
```
